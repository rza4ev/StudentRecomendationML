#  Model Training

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Veri çerçevenizi yükleyin (veriyi burada elde etmeniz gerekecektir)
# Örnek olarak, "X" bağımsız değişkenlerinizi ve "y" bağımlı değişkeninizi tanımlayın.
# X, bağımsız değişkenler
# y, bağımlı değişken

# Örnek veriyi kullanarak, bağımsız değişkenleri ve bağımlı değişkeni tanımlayalım
X = df.drop(columns=['Specialization'])  # 'Specialization' sütununu çıkartarak bağımsız değişkenleri oluşturur
y = df['Specialization']  # Bağımlı değişkeni 'Specialization' olarak tanımlar

# Veriyi eğitim ve test setlerine bölelim
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X)
```

```python
df.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier

# RandomForestClassifier modelini oluşturun ve hiperparametreleri ayarlayın
rf_classifier = RandomForestClassifier(
    n_estimators=100,  
    max_depth=11,      
    min_samples_split=5,  
    min_samples_leaf=5,   
    max_features='auto',  
    random_state=42       
)

# Modeli eğitin
rf_classifier.fit(X_train, y_train)



```

```python
# Test verilerini kullanarak tahminler yapın
y_pred = rf_classifier.predict(X_test)

# Sınıflandırma sonuçlarını değerlendirin
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", report)

```

```python
from sklearn.metrics import accuracy_score

# Train seti için tahminler yapın
y_train_pred = rf_classifier.predict(X_train)

# Train seti doğruluk skorunu hesaplayın
accuracy_train = accuracy_score(y_train, y_train_pred)

# Test seti için tahminler yapın
y_test_pred = rf_classifier.predict(X_test)

# Test seti doğruluk skorunu hesaplayın
accuracy_test = accuracy_score(y_test, y_test_pred)

# Sonuçları görüntüleyin
print("Train Seti Doğruluk Skoru:", accuracy_train)
print("Test Seti Doğruluk Skoru:", accuracy_test)
```
