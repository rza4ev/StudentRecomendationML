```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Veri Ã§erÃ§evenizi yÃ¼kleyin (veriyi burada elde etmeniz gerekecektir)
# Ã–rnek olarak, "X" baÄŸÄ±msÄ±z deÄŸiÅŸkenlerinizi ve "y" baÄŸÄ±mlÄ± deÄŸiÅŸkeninizi tanÄ±mlayÄ±n.
# X, baÄŸÄ±msÄ±z deÄŸiÅŸkenler
# y, baÄŸÄ±mlÄ± deÄŸiÅŸken

# Ã–rnek veriyi kullanarak, baÄŸÄ±msÄ±z deÄŸiÅŸkenleri ve baÄŸÄ±mlÄ± deÄŸiÅŸkeni tanÄ±mlayalÄ±m
X = df.drop(columns=['Specialization'])  # 'Specialization' sÃ¼tununu Ã§Ä±kartarak baÄŸÄ±msÄ±z deÄŸiÅŸkenleri oluÅŸturur
y = df['Specialization']  # BaÄŸÄ±mlÄ± deÄŸiÅŸkeni 'Specialization' olarak tanÄ±mlar

# Veriyi eÄŸitim ve test setlerine bÃ¶lelim
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X)
```

```python
df.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier

# RandomForestClassifier modelini oluÅŸturun ve hiperparametreleri ayarlayÄ±n
rf_classifier = RandomForestClassifier(
    n_estimators=100,  
    max_depth=11,      
    min_samples_split=5,  
    min_samples_leaf=5,   
    max_features='auto',  
    random_state=42       
)

# Modeli eÄŸitin
rf_classifier.fit(X_train, y_train)



```

```python
# Test verilerini kullanarak tahminler yapÄ±n
y_pred = rf_classifier.predict(X_test)

# SÄ±nÄ±flandÄ±rma sonuÃ§larÄ±nÄ± deÄŸerlendirin
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", report)

```

```python
from sklearn.metrics import accuracy_score

# Train seti iÃ§in tahminler yapÄ±n
y_train_pred = rf_classifier.predict(X_train)

# Train seti doÄŸruluk skorunu hesaplayÄ±n
accuracy_train = accuracy_score(y_train, y_train_pred)

# Test seti iÃ§in tahminler yapÄ±n
y_test_pred = rf_classifier.predict(X_test)

# Test seti doÄŸruluk skorunu hesaplayÄ±n
accuracy_test = accuracy_score(y_test, y_test_pred)

# SonuÃ§larÄ± gÃ¶rÃ¼ntÃ¼leyin
print("Train Seti DoÄŸruluk Skoru:", accuracy_train)
print("Test Seti DoÄŸruluk Skoru:", accuracy_test)
```
